```python
import pandas as pd
df = pd.read_csv('HR_Analysis_cleaned.csv')
```

```python
display(df.head())
df.info()
display(df.isnull().sum())
```

```python
display(df.describe())

for col in df.columns:
    if df[col].dtype == 'object':
        print(f"Value counts for column: {col}")
        display(df[col].value_counts())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Histogram for 'Age'
plt.figure(figsize=(10, 6))
sns.histplot(df['Age'], kde=True)
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

# 2. Bar plot for 'Department'
plt.figure(figsize=(12, 6))
sns.countplot(data=df, y='Department', order=df['Department'].value_counts().index)
plt.title('Distribution of Employees Across Departments')
plt.xlabel('Number of Employees')
plt.ylabel('Department')
plt.show()

# 3. Scatter plot of 'Years_At_Company' vs 'Monthly_Salary'
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Years_At_Company', y='Monthly_Salary', alpha=0.5)
plt.title('Relationship Between Years at Company and Monthly Salary')
plt.xlabel('Years at Company')
plt.ylabel('Monthly Salary')
plt.show()

# 4. Box plot of 'Employee_Satisfaction_Score' by 'Resigned'
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='Resigned', y='Employee_Satisfaction_Score')
plt.title('Employee Satisfaction Score by Resignation Status')
plt.xlabel('Resigned (True/False)')
plt.ylabel('Employee Satisfaction Score')
plt.show()
```

## 1 Data Overview
Before visualizations, it's important to understand the data structure and summary insights.

```python

# Basic Data Overview
df.info()

# Summary Statistics
df.describe(include='all')

# Missing Values Check
df.isnull().sum()

# Duplicate Check
df.duplicated().sum()
```

## 2 Demographic Analysis

```python

import seaborn as sns
import matplotlib.pyplot as plt

# Gender Distribution
sns.countplot(x='Gender', data=df)
plt.title('Gender Distribution')
plt.show()

# Education Field Distribution
sns.countplot(x='Education_Level', data=df)
plt.title('Distribution of Education Fields')
plt.xticks(rotation=45)
plt.show()

# Marital Status vs Attrition
sns.countplot(x='MaritalStatus', hue='Attrition', data=df)
plt.title('Marital Status vs Attrition')
plt.show()
```

## 3 Job & Salary Insights

```python

# Monthly Income Distribution
sns.histplot(df['MonthlyIncome'], kde=True, bins=30)
plt.title('Monthly Income Distribution')
plt.show()

# Income by Department
sns.boxplot(x='Department', y='MonthlyIncome', data=df)
plt.title('Monthly Income by Department')
plt.xticks(rotation=45)
plt.show()

# Job Role vs Monthly Income
sns.boxplot(x='JobRole', y='MonthlyIncome', data=df)
plt.title('Job Role vs Monthly Income')
plt.xticks(rotation=45)
plt.show()
```

## 4 Correlation & Relationships

```python

# Correlation Heatmap
corr = df.corr(numeric_only=True)
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()

# Attrition vs Years at Company
sns.boxplot(x='Attrition', y='YearsAtCompany', data=df)
plt.title('Attrition vs Years at Company')
plt.show()
```

## 5 Attrition & Satisfaction Deep Dive

```python

# Attrition by Department
sns.countplot(x='Department', hue='Attrition', data=df)
plt.title('Attrition by Department')
plt.xticks(rotation=45)
plt.show()

# Attrition by Overtime
sns.countplot(x='OverTime', hue='Attrition', data=df)
plt.title('Attrition by Overtime')
plt.show()

# Attrition by Job Satisfaction
sns.boxplot(x='Attrition', y='JobSatisfaction', data=df)
plt.title('Attrition by Job Satisfaction')
plt.show()
```

## 6 Additional Useful Visualizations

```python

# Age vs Salary Scatter
sns.scatterplot(x='Age', y='MonthlyIncome', hue='JobLevel', data=df)
plt.title('Age vs Monthly Income by Job Level')
plt.show()

# Years at Company vs Percent Salary Hike
sns.lineplot(x='YearsAtCompany', y='PercentSalaryHike', data=df)
plt.title('Years at Company vs Percent Salary Hike')
plt.show()

# Employee Count by Business Travel
sns.countplot(x='BusinessTravel', data=df)
plt.title('Employee Count by Business Travel Frequency')
plt.show()
```


## 7 Summary & Insights

- Most employees are aged between **2555 years**.
- **Departments** have nearly uniform employee distribution.
- **Monthly income** positively correlates with **job level** and **years at company**.
- **Attrition** is slightly higher among employees who do **overtime** and have **lower satisfaction scores**.
- Management roles earn significantly higher than entry-level positions.
- The dataset shows a generally balanced demographic distribution.

>  This enhanced EDA provides a comprehensive and professional view of HR analytics, perfect for reporting and presentation.

```python
df["Hire_Year"] = pd.to_datetime(df["Hire_Date"]).dt.year
df["Hire_Month"] = pd.to_datetime(df["Hire_Date"]).dt.month
df["Hire_Month&Year"] = (df["Hire_Year"].astype(str)+ "-" + df["Hire_Month"].astype(str).str.zfill(2)).astype('category')
```

```python

plt.figure(figsize = (18,8))
sns.lineplot(data = df, x = "Hire_Month&Year", y = "Monthly_Salary", ci = None, linewidth=2, color="#007ACC")
plt.xticks(rotation = 45, fontsize = 13)
import matplotlib.dates as mdates

plt.gca().xaxis.set_major_locator(mdates.MonthLocator(interval=5))
plt.xlabel("Hire Month & Year", fontsize=15, labelpad=10)
plt.ylabel("Monthly Salary", fontsize=15, labelpad=10)
plt.title("Monthly Salary Trend Over Time", fontsize=18, pad=15)
sns.despine()
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
Dept = ['IT', 'Finance', 'Customer Support', 'Engineering', 'Marketing',
       'HR', 'Operations', 'Sales', 'Legal']

plt.figure(figsize = (12,12))
for i,col in enumerate(Dept,1):
      plt.subplot(5,2,i+1)
      sns.lineplot(
            data = df[df["Department"] == col],
            x = "Hire_Month&Year",
            y = "Employee_ID",
            estimator = "count",
            ci = None,
            linewidth = 2)
      plt.xlabel("Years and Months")
      plt.ylabel("Counts")
      plt.title(f"Hire Emp counts of Dept {col} by Time")
plt.tight_layout()
plt.show()
```
