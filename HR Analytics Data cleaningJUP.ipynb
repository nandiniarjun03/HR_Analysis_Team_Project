```python
# importing libraries

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
pd.set_option("display.max_column",None)
```

```python
data = pd.read_csv("HR Analytics Dataset.csv")
d = pd.DataFrame(data)
df = d.copy()
```

```python
df["Hire_Date"] = pd.to_datetime(df["Hire_Date"]).dt.date
df["Hire_Date"]  = pd.to_datetime(df["Hire_Date"] )
```

```python
cat_col = ['Employee_ID','Department', 'Gender', 'Job_Title',
        'Education_Level', 'Remote_Work_Frequency','Resigned']
for col in cat_col:
    df[col] = df[col].astype('category')
```

```python
df.head(2)
```

```python
# Let's update the Years_At_Company column based on today date
from datetime import datetime
df["Years_At_Company"] = ((datetime.now() - df["Hire_Date"]).dt.days/365).astype('int')
```

```python
# lets check dupliactes rows in Employee_ID column
df["Employee_ID"].duplicated().sum()
```

### No duplicated rows in Employee_ID column

```python
# Lets see is here any mistyped value in category columns?
category_col = ['Department', 'Gender', 'Job_Title',
        'Education_Level', 'Remote_Work_Frequency','Resigned']

plt.figure(figsize = (16,10))


for i,col in enumerate (category_col):
    plt.subplot(3,2,i+1)
    df[col].value_counts().plot(kind = "bar")


plt.tight_layout()
plt.show()
```

```python
# Now correctly renaming the values of Remote_Work_Frequency for better understanding

remapping_values = {0 : "Never", 25 : "Occasionally", 50 : "Half-time", 75 : "Frequently", 100 : "Fully Remote"}
df["Remote_Work_Frequency"] = df["Remote_Work_Frequency"].map(remapping_values)
```

```python
df.head(2)
```

```python
# Now we are going to check the data distribution to find out the outliers
df.info()
```

```python
num_col = ['Age', 'Years_At_Company', 'Performance_Score', 'Work_Hours_Per_Week',
       'Projects_Handled', 'Overtime_Hours', 'Sick_Days', 'Team_Size',
       'Training_Hours', 'Promotions']

plt.figure(figsize = (14,8))

for i,col in enumerate (num_col):
    plt.subplot(5,2,i+1)
    sns.violinplot(data = df, x = col)

plt.tight_layout()
plt.show()
```
